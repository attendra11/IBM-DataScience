```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

def scrape_gme_revenue():
    # URL for GameStop revenue data
    url = 'https://www.macrotrends.net/stocks/charts/GME/gamestop/revenue'

    # Send a GET request to the URL
    response = requests.get(url)

    # Create a BeautifulSoup object from the response content
    soup = BeautifulSoup(response.content, 'html.parser')

    # Find the revenue table on the page
    table = soup.find('table', {'class': 'historical_data_table'})

    # Extract the table data into a pandas dataframe
    gme_revenue = pd.read_html(str(table))[0]

    return gme_revenue

# Call the function to scrape GME revenue data
gme_revenue = scrape_gme_revenue()

# Display the last five rows
print(gme_revenue.tail())
```

```python
import matplotlib.pyplot as plt
import yfinance as yf

def make_graph(data, title):
    # Create a figure and axis
    fig, ax = plt.subplots()

    # Reset the index
    data = data.reset_index()

    # Plot the stock data
    ax.plot(data.index, data['Close'])

    # Set the title
    ax.set_title(title)

    # Rotate x-axis labels for better readability
    plt.xticks(rotation=45)

    # Display the graph
    plt.show()

# Fetch Tesla stock data
tesla_data = yf.download('TSLA', start='2023-01-01', end='2023-05-31')

# Call the make_graph function to plot the Tesla stock data
make_graph(tesla_data, title='Tesla Stock Data')
```
