```python
import pandas as pd
from sqlalchemy import create_engine
import sqlite3
```

```python
census_data_df = pd.read_csv('ChicagoCensusData.csv')
chicago_public_schools_df = pd.read_csv('chicago_public_schools.csv')
chicago_crime_data_df = pd.read_csv('ChicagoCrimeData.csv')
```

```python
engine = create_engine('sqlite:///FinalDB.db')
```

```python
census_data_df.to_sql('ChicagoCensusData', engine, if_exists='replace', index=False)
chicago_public_schools_df.to_sql('chicago_public_schools', engine, if_exists='replace', index=False)
chicago_crime_data_df.to_sql('ChicagoCrimeData', engine, if_exists='replace', index=False)
```

### Problem 1

##### Find the total number of crimes recorded in the CRIME table.


```python
import sqlite3

# Connect to the SQLite database
conn = sqlite3.connect('FinalDB.db')
cursor = conn.cursor()

# Execute the SQL query to count the number of crimes
cursor.execute('SELECT COUNT(*) FROM ChicagoCrimeData')
total_crimes = cursor.fetchone()[0]

# Print the total number of crimes
print("Total number of crimes recorded:", total_crimes)

# Close the database connection
cursor.close()
conn.close()
```

### Problem 2

##### List community areas with per capita income less than 11000.


```python
import sqlite3

# Connect to the SQLite database
conn = sqlite3.connect('FinalDB.db')
cursor = conn.cursor()

# Execute the SQL query to select community areas with per capita income less than 11000
cursor.execute('SELECT COMMUNITY_AREA_NAME FROM ChicagoCensusData WHERE PER_CAPITA_INCOME < 11000')
community_areas = cursor.fetchall()

# Print the community areas with per capita income less than 11000
print("Community areas with per capita income less than 11000:")
for area in community_areas:
    print(area[0])

# Close the database connection
cursor.close()
conn.close()
```

### Problem 9

##### Use a sub-query to find the name of the community area with highest hardship index


```python
import sqlite3

# Connect to the SQLite database
conn = sqlite3.connect('FinalDB.db')
cursor = conn.cursor()

# Execute the SQL query with a subquery to find the community area with the highest hardship index
cursor.execute('SELECT COMMUNITY_AREA_NAME FROM ChicagoCensusData WHERE HARDSHIP_INDEX = (SELECT MAX(HARDSHIP_INDEX) FROM ChicagoCensusData)')
community_area = cursor.fetchone()

# Print the community area with the highest hardship index
print("Community area with the highest hardship index:", community_area[0])

# Close the database connection
cursor.close()
conn.close()
```

### Problem 10

##### Use a sub-query to determine the Community Area Name with most number of crimes?


```python
import sqlite3

# Connect to the SQLite database
conn = sqlite3.connect('FinalDB.db')
cursor = conn.cursor()

# Execute the SQL query with a subquery to find the community area name with the most number of crimes
cursor.execute('SELECT COMMUNITY_AREA_Number FROM ChicagoCrimeData WHERE ID = (SELECT ID FROM ChicagoCrimeData GROUP BY COMMUNITY_AREA_Number ORDER BY COUNT(*) DESC LIMIT 1)')
community_area = cursor.fetchone()

# Print the community area name with the most number of crimes
print("Community area with the most number of crimes:", community_area[0])

# Close the database connection
cursor.close()
conn.close()
```
